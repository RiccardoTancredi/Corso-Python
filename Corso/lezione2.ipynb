````python
# magic method
%matplotlib inline  

import numpy as np
import matplotlib.pyplot as plt
````

## Istogrammi

````python
N = 1000
dati = np.random.normal(size=N)
plt.hist(dati)
````

````python
n_bins = 15 # int(np.sqrt(N))
plt.hist(dati, bins=n_bins, color='lightblue', 
         label='Dati', density=True)
plt.xlabel('Dati')
plt.ylabel('Frequenze relative')
plt.legend()
plt.grid(ls='--', alpha=0.5)
plt.show()
````

* Come posso fare un plot sopra un altro plot?  possibile?

````python
def gaussian(x, loc=0, scale=1):
    return np.exp(-0.5*((loc-x)/scale)**2)/np.sqrt(2*np.pi*scale**2)

xx = np.linspace(-3.5, 3.5, N)
plt.hist(dati, bins=n_bins, color='lightblue', 
         label='Dati', density=True)
plt.plot(xx, gaussian(xx), label=r'$\mathcal{N}(0, 1)$', color='red')
plt.xlabel('Dati')
plt.ylabel('Frequenze relative')
plt.legend(frameon=False, fontsize=15)
plt.title('Istogramma + Gaussiana', fontsize=15)
plt.grid(ls='--', alpha=0.5)
plt.show()
````

1. Esercizio: 
    * Genera dei dati distribuiti esponezialmente (per sempliciti puoi lasciare $\lambda = 1 = 1/\beta$);
    * Sovrimponi una curva esponeziale di colore viola, tratteggiata e con spessore della linea pari a **3**;
    * Cambia la scala di visualizzazione in logaritmica in modo tale che la curva precedente diventi una retta: 
        * cerca come si usano i metodi `plt.xscale()` e/o `plt.yscale()`.

````python
dati = np.random.exponential(size=N)
plt.hist(dati, bins=50, density=True, 
         label='Dati', color='royalblue', edgecolor='white')
xx = np.linspace(0, max(dati), N)
plt.plot(xx, np.exp(-xx), label=r'$\lambda\cdot e^{-x\cdot\lambda}$', 
         color='purple', ls='--', lw=3)
plt.xlabel('Dati')
plt.ylabel('Frequenze relative')
plt.legend(frameon=False, fontsize=15)
plt.title('Istogramma + Esponenziale', fontsize=15)
plt.grid(ls=':', alpha=0.5)
# plt.yscale('log')
plt.show()
````

Possiamo fare due plot in un unico grafico

````python
r=[1,6,5,13]
s=[0.5,4,6,3]
x=[3,9,8,7]
y=[2.6,4,6,3]

plt.plot(x,y, "*y", label="Stelle")
plt.plot(r,s, "or", label="Fragole")

plt.legend(loc="best")
plt.show()
````

Aggiungiamo informazioni al grafico

````python
r=[1,6,5,13]
s=[0.5,4,6,3]
x=[3,9,8,7]
y=[2.6,4,6,3]

plt.plot(x,y, "*y", label="Stelle")
plt.plot(r,s, "or", label="Fragole")

plt.legend(loc="best")
plt.xlabel("spazio [m]")
plt.ylabel("velocit [$m\;s^{-1}$]")
plt.xlim(2.7,9.3)
plt.ylim(2.4,6.2)

plt.title("Esempio di grafico")
plt.show()
````

Aggiungiamo barre d'errore alle fragole e modifichiamo la dimensione del marker delle stelle

````python
r=[1,6,5,13]
s=[0.5,4,6,3]
x=[3,9,8,7]
y=[2.6,4,6,3]

plt.plot(x,y, "*y", label="Banane", markersize="10")
plt.errorbar(r,s, xerr=0.3,yerr=0.2, marker='o', color='r', linewidth=0, capsize=3, elinewidth=1, label="Fragole")

plt.legend(loc="best")
plt.xlabel("spazio [m]")
plt.ylabel("velocit [$m\;s^{-1}$]")
plt.xlim(2.7,9.3)
plt.ylim(2.4,6.2)

plt.title("Esempio di grafico")
plt.show()
````

Mettiamo una griglia e modifichiamo i font dei label e del titolo

````python
r=[1,6,5,13]
s=[0.5,4,6,3]
x=[3,9,8,7]
y=[2.6,4,6,3]

plt.plot(x,y, "*y", label="Banane", markersize="10")
plt.errorbar(r,s, xerr=0.3,yerr=0.2, marker='o', color='r', linewidth=0, capsize=3, elinewidth=1, label="Fragole")

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.legend(loc="best")
plt.xlabel("spazio [m]", fontdict=font2)
plt.ylabel("velocit [$m\;s^{-1}$]", fontdict=font2)
plt.xlim(2.7,9.3)
plt.ylim(2.4,6.2)
#plt.yscale('log')

plt.title("Esempio di grafico", fontdict=font1)
plt.grid(True, alpha=0.6, linestyle=':')
plt.show()
````
